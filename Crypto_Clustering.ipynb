# Crypto Clustering

```python
# load dependencies
import pandas as pd
import hvplot.pandas
from sklearn.cluster import KMeans
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
```

## Data Loading & Familiarization

```python
# import crypto market data into a data frame
filepath = 'Resources/crypto_market_data.csv'
raw_cm_df = pd.read_csv(filepath, index_col='coin_id')
raw_cm_df.head(5)
```

```python
# visualize summary statistics to get a high-level understanding of the data
raw_cm_df.describe()
```

```python
raw_cm_df.hvplot.line(width=800,
                      height=400,
                      rot=90)
```

## Data Preparation

```python
# use the `StandardScaler()` module from `scikit-learn` to normalize the data from the CSV file
scaler = StandardScaler()
scaler.fit(raw_cm_df)
scaled_cm_df = scaler.transform(raw_cm_df)

# visualize
scaled_cm_df
```

```python
# convert the scaled output back to a data frame
scaled_df = pd.DataFrame(scaled_cm_df, index=raw_cm_df.index, columns=raw_cm_df.columns)
scaled_df.head(5)
```

## Find the Best Value for $k$ Using the Original Scaled DataFrame

```python
# instantiate an empty list to record inertia and a list of k values, 1 through 11
inertia = []
k_list = list(range(1, 12))

# create a for loop where each value of k is evaluated using the K-means algorithm
for k in k_list:

    # create a KMeans model using the loop counter for the n_clusters
    k_model = KMeans(n_clusters=k, random_state=1, n_init=10)

    #fit the model to the scaled data
    k_model.fit(scaled_df)

    # append the model.inertia_ to the inertia list
    inertia.append(k_model.inertia_)

# create a dictionary that holds the list values for k and inertia
elbow_data = {"k": k_list, "inertia": inertia}

# create a data frame using the elbow_data dictionary
df_elbow = pd.DataFrame(elbow_data)

# plot the data frame
orig_elbow = df_elbow.hvplot.line(
    x="k", 
    y="inertia", 
    title="Oiringal Elbow Curve", 
    xticks=k
    )

orig_elbow
```

The slope of the elbow curve significantly flattens at $k=4$, which tells us this is likely our "ideal" value for $k$.

## Cluster Cryptocurrencies with $k$-means Using the Original Scaled Data

```python
# initialize the KMeans model using the best value of k
# use a random_state of 1 to generate the model
model = KMeans(n_clusters=4, random_state=1, n_init=10)

# fit the KMeans model using the scaled data
model.fit(scaled_df)

# predict the clusters to group the cryptocurrencies using the scaled data
clusters = model.predict(scaled_df)

# print the resulting array of cluster values
print(clusters)
```

```python
# create a copy of the data frame
predictions_df = scaled_df.copy()

# add a new column to the data frame with the predicted clusters
predictions_df['clusters'] = clusters

# Display sample data
predictions_df.head(5)
```

```python
# create a scatter plot using hvPlot by setting 
# add the crypto name in the `hover_cols` parameter to identify the cryptocurrency represented by each data point
orig_clusters = predictions_df.hvplot.scatter(
    title="Original Clustering",
    x="price_change_percentage_24h",
    y="price_change_percentage_7d",
    by="clusters",
    hover_cols = ["coin_id"],
    ).opts(yformatter="%.0f")

orig_clusters
```

## Optimize Clusters with Principal Component Analysis

```python
# create a PCA model instance and set `n_components=3`
pca = PCA(n_components=3)
```

```python
# use the PCA model with `fit_transform` to reduce to three principal components
crypto_pca = pca.fit_transform(scaled_df)

# view the first five rows of the data frame
crypto_pca[:5]
```

```python
# retrieve the explained variance to determine how much information can be attributed to each principal component
sum(pca.explained_variance_ratio_)
```

About 89.5% of the total explained variance of the total variance is condensed into the three principal components.

```python
# create a new data frame with the PCA data
crypto_pca_df = pd.DataFrame(crypto_pca,
                             columns=['PC1', 'PC2', 'PC3'])

# copy the crypto names from the original data
coin_id = raw_cm_df.index

# set the coinid column as index
crypto_pca_df['coin_id'] = coin_id.values
crypto_pca_df = crypto_pca_df.set_index('coin_id')

# display sample data
crypto_pca_df.head(5)
```

## Find the Best Value for $k$ Using the PCA Data

```python

# create an empy list to store the inertia values
inertia = []

# create a list with the number of k-values from 1 to 11
k_list = list(range(1, 12))

# create a for loop where each value of k is evaluated using the K-means algorithm
for k in k_list:

    # create a KMeans model using the loop counter for the n_clusters
    k_model = KMeans(n_clusters=k, random_state=1, n_init=10)

    #fit the model to the pca data
    k_model.fit(crypto_pca_df)

    # append the model.inertia_ to the inertia list
    inertia.append(k_model.inertia_)

# create a dictionary that holds the list values for k and inertia
elbow_data = {'k' : k_list, 'inertia' : inertia}

# create a data frame using the elbow_data dictionary
df_elbow = pd.DataFrame(elbow_data)

# plot the data frame
pca_elbow = df_elbow.hvplot.line(
    x='k',
    y='inertia',
    title='PCA Elbow Curve',
    xticks=k
    )

pca_elbow
```

The best value for $k$ when using the PCA data is 4. Note that this _does not differ_ from the best $k$ value found using the original scaled data.

## Cluster Cryptocurrencies with $k$-means Using the PCA Data

```python
# initialize the KMeans model using the best value for k
pca_model = KMeans(n_clusters=4, random_state=1, n_init=10)

# fit the KMeans model using the pca data
pca_model.fit(crypto_pca_df)

# predict the clusters to group the cryptocurrencies using the pca data
pca_clusters = pca_model.predict(crypto_pca_df)

# print the resulting array of cluster values
print(pca_clusters)
```

```python
# create a copy of the data frame with the pca data
crypto_predictions_df = crypto_pca_df.copy()

# add a new column to the data frame with the predicted clusters
crypto_predictions_df['clusters'] = pca_clusters

# display sample data
crypto_predictions_df.head(5)
```

```python
# create a scatter plot using hvPlot
# add the crypto name in the `hover_cols` parameter to identify
# the cryptocurrency represented by each data point
pca_clusters = crypto_predictions_df.hvplot.scatter(
    title="PCA Clustering",
    x="PC1",
    y="PC2",
    by="clusters",
    hover_cols = ["coin_id"],
    ).opts(yformatter="%.0f")

pca_clusters
```

## Visualize and Compare the Results

```python
# composite plot to contrast the Elbow curves
(orig_elbow + pca_elbow).cols(1)
```

```python
# Composite plot to contrast the clusters
(orig_clusters + pca_clusters).cols(1)
```

While the scatterplot representation of the clusters appears to have changed drastically, it seems that we have approximately the same distribution of data points within each cluster, with similar separation as represented in a 2-dimensional plane. This, along with the variance capture measures examined above, indicates that condensing the data through principal component analysis has not resulted in significant information loss.
